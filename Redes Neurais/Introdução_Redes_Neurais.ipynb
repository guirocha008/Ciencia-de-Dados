$$Conceitos:$$

**Redes Neurais** =   São técnicas computacionais que apresentam um modelo matemático inspirado na estrutura neuronal  de  organismos  inteligentes e  que  adquirem conhecimento através da experiência.



- **Deep Learning** = Arquitetura da Rede Neural


- **Função de Ativação** = Processa o sinal gerado pela combinação linear das entradas e dos pesos das sinapses, para gerar o sinal de saída do neurônio, podendo ter resposta lineares e não-lineares 


- **Épocas** = Quantas vezes os dados passam pela rede



- **Early Stopping** = Melhoria não está mais ocorrendo? Pare!, técnica para evitar o overfitting.


- **Dropout** = Diminui a capacidade de memorização do neurônio excluindo aleatoriamente alguns neurônios de algumas camadas, evitando overfitting


- **Arquiteturas esparsas** = São arquiteturas com menos conexões entre camadas


- **Camada** = Cada camada criar uma separação entre os dados, quando mais camadas mais os dados estão propensos a overfitting


- **Camada Densa** = Conecta cada neurônio em uma camada com todos os neurônios da camada anterior


- **Descida do gradiente** = Processo utilizado para encontrar o mínimo da função custo com base em variação nos pesos, através do processo de backpropagation os erros obtidos após os cálculos dos pesos são retroalimentados na rede até otimizarmos o resultado


**Otimizadores:**

- **Descida do gradiente** = Taxa de aprendizado fixa


- **ADAM, RMSPROP, ADAGRAD** = Taxa de aprendizado que varia com as iterações




**Redes neurais rasas (shallow):**

-  Possui apenas uma camada oculta, trata apenas de dados linearmente separáveis


- **Perceptron (P), Feed Forward Network (FFN), Radial Basis Network (RBF)** = São o tipo mais básico, onde a informação de entrada flui em sequência linear até a saída.


**Redes neurais profundas:** 

- Quando as redes neurais possuem muitas camadas ocultas (deep learning), trata-se de dados não linearmente separáveis.




**Função de Custo:** 

- Função que mensura o quanto estamos acertando


- São diferentes para variáveis contínuas e categóricas


- **Contínuas:** MSE, MAE


- **Categoricas:** Binary cross-entropy, Categorical cross-entropy


**Viés x Variância:**


- Ausência de Viés - na média você acerta.


- Redução de Variância – tiro ao alvo


```python
#Importar as bibliotecas
import numpy as np
import pandas as pd
```

## Carregando o dataset

```python
#Carregar dataset
dados = pd.read_csv('winequality-red.csv',encoding='latin1',sep=',')
dados
```

```python
#Tipos de dados
dados.info()
```

```python
#Separar os dados em X e Y
x = dados.iloc[:,0:11]
y = dados.quality

#Padronização dos dados

dados_padro = (x - x.mean()) / (x.std())
dados_padro
```

```python
#Divisão dos dados 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dados_padro,y,test_size=0.3,random_state=1)

#Alterar as dimensoes
X_train = X_train.T
X_test = X_test.T
y_train = pd.DataFrame(y_train).T
y_test = pd.DataFrame(y_test).T
```

**As redes neurais feedforward avançam da esquerda para direita, por isso fazemos a transposição dos dados para deixar as variáveis em linhas**

# Parte 1 - Criar arquitetura da rede

```python
#Arquitetura
def arq_func(x,y,hidden_neurons):
    n_x = x.shape[0] #quantidade de linhas de x = neurônios da camada de entrada
    n_h = hidden_neurons #quantidade de neurônios na camada escondida
    n_y = y.shape[0] #quantidade de linhas de y = neurônios da camada de saída
    arquite = {'n_x': n_x, 'n_h': n_h, 'n_y': n_y}
    return(arquite)
    
    
#Aplicar a função arquitetura de rede
arq_rede = arq_func(X_train, y_train , 4)
arq_rede
```

**Na primeira etapa, criamos a arquitetura da rede, passando os valores das variáveis de entrada, camada oculta e saída**

# Parte 2 - Inicializar os Parâmetros randomicamente

```python
#Inicializar pesos
from random import randrange, uniform

def init_pesos(x, arquite):
    n_x = arquite.get('n_x')
    n_h = arquite.get('n_h')
    n_y = arquite.get('n_y')
    
    W1 = np.full((n_h, n_x), np.random.rand(n_h, n_x))*0.01
    W2 = np.full((n_y, n_h), np.random.rand(n_y, n_h))*0.01
    
    params_pesos = {'W1': W1, 'W2': W2}
    return(params_pesos)
    
    
    
inic_pesos = init_pesos(X_train,arq_rede)
inic_pesos
```

**A partir da arquitetura, inicializamos os parâmetros (pesos e bias) randomicamente. O primeiro conjunto são os pesos da camada de entrada com a camada intermediaria. O segundo são os pesos da camada intermediaria com a camada de saída**

# Parte 3 - Função de ativação - Sigmoide

```python
def sig(x):
    return((1/(1+np.exp(-x))))
```

# Parte 4 - Forward Propagation

```python
def forward_prop(x,params_pesos,arquite):
    
    n_h = arquite.get('n_h')
    n_y = arquite.get('n_y')
    
    W1 = params_pesos.get('W1')
    W2 = params_pesos.get('W2')
    
    
    Z1 = np.dot(W1,x) #Multiplicação entre camada de entrada X Pesos
    A1 = sig(Z1) #Função Sigmoide do resultado da Multiplicação entre Camada de entrada X Pesos (Camada Oculta)
    Z2 = np.dot(W2,A1) #Multiplicação entre o resultado da sigmoide X pesos da camada intermediaria (Entre camada oculta e saida)
    prev1 = sig(Z2) #Função Sigmoide no resultado da multiplicação de Z2 (Camada de saida)
    
    cache = {'Z1':Z1, 'A1':A1, 'Z2':Z2, 'prev1':prev1}

    return (cache)
    
    
forward_1 = forward_prop(X_train,inic_pesos,arq_rede)
forward_1
```

**Agora com todas as informações dos pesos e estrutura da rede, iniciamos o processo de geração de resultados, multiplicação das matrizes (Variáveis e pesos)**

# Parte 5- Cost Function

```python
def funcao_custo(y, cache):
    
    
    m = y.shape[1]
    
    y = y.values[0]
    
    prev1 = cache.get('prev1')[0]

    cost = sum((y-prev1)**2)/m
    
    return (cost)


custo_1 = funcao_custo(y_train,forward_1)
custo_1
```

**Função de custo: Mean Squared Error, função que vai calcular o quanto a rede está acertando**

# Parte 6 - Backpropagation

```python
def backwardPropagation(x, y, cache, params_pesos, arquite):
    
    m = x.shape[1]
    
    n_x = arquite.get('n_x')
    n_h = arquite.get('n_h')
    n_y = arquite.get('n_y')

    prev1 = cache.get('prev1')[0]
    A1 = cache.get('A1')
    W2 = params_pesos.get('W2')

    
    dZ2 = prev1 - y #Valor previsto - Valor real (Diferença para alterar os pesos)
    dW2 = 1/m * np.dot(dZ2, A1.T) #Quanto eu preciso alterar os pesos de W2

    
    dZ1 = np.dot(W2.T,dZ2) * (1 - A1**2)#Mudança que ocorre na camada intermediaria para camada de entrada em funçao de DZ2
    dW1 = 1/m * np.dot(dZ1 , x.T) #Quanto eu preciso alterar os pesos de W1 em função de W2

    
    grads = {'dW1':dW1, 'dW2':dW2}
   
    
    return(grads)


gradiente = backwardPropagation(X_train,y_train,forward_1,inic_pesos,arq_rede)
gradiente
```

**Após o cálculo da função de custo, é feito o ajuste dos pesos, o backpropagation dá a direção (valor positivo ou negativo) e o valor para que seja atualizado os pesos**

# Parte 7 - Atualizar os pesos

```python
def updateParameters(grads, params_pesos, learning_rate):
    
    W1 = params_pesos.get('W1')
    W2 = params_pesos.get('W2')
    
    dW1 = grads.get('dW1')
    dW2 = grads.get('dW2')
    
    
    W1 = W1 - learning_rate * dW1 #Atualizar peso W1
    W2 = W2 - learning_rate * dW2 #Atualizar peso W2
    
    updated_pesos = {'W1': W1, 'W2': W2}
    
    
    return (updated_pesos)


atualiza_pesos = updateParameters(gradiente,inic_pesos,learning_rate = 0.1)
atualiza_pesos
```

**Com o valor do backpropagation é feita atualização dos pesos, para que na próxima iteração o valor do custo seja menor, por isso diminuímos o valor do peso atual pelo peso adquirido pelo backpropagation, não podendo esquecer da taxa de aprendizado**

# Parte 9 - Treinar o modelo

```python
def trainModel(x, y, num_iteration, hidden_neurons, learning_rate):
    
    arq_rede = arq_func(x, y, hidden_neurons)
    init_params = init_pesos(x, arq_rede)
    
    cost_history = []
    

    for i in range(1,num_iteration):
        fwd_prop = forward_prop(x, init_params, arq_rede)
        cost = funcao_custo(y, fwd_prop)
        back_prop = backwardPropagation(x, y, fwd_prop, init_params, arq_rede)
        update_params = updateParameters(back_prop, init_params, learning_rate = learning_rate)
        init_params = update_params
        cost_history.append(cost)
        
        
        
    
    
    model_out = {'cost_hist': cost_history}


    return (model_out)


modelo_final = trainModel(X_train,y_train,num_iteration=100,hidden_neurons=20,learning_rate=0.01)
modelo_final
```

**Loop de todo o processo**
