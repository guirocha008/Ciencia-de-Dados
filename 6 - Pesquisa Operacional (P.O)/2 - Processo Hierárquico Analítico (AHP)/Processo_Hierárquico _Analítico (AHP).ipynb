-----------------------------------------------------------------------------------------------------------------------



**Tomada de Decis√£o:**

    

    

Uma decis√£o precisa ser tomada sempre que estamos diante de um problema que possui mais de uma alternativa para sua solu√ß√£o. Mesmo quando, para solucionar um problema, possu√≠mos uma √∫nica a√ß√£o a tomar, temos as alternativas de tomar ou n√£o essa a√ß√£o



O processo de decis√£o requer a exist√™ncia de um conjunto de alternativas fact√≠veis para sua composi√ß√£o, em que cada decis√£o tem associados um ganho e uma perda



-----------------------------------------------------------------------------------------------------------------------



**M√©todos Multicrit√©rio:**



Problemas de tomada de decis√£o multicrit√©rio s√£o caracterizados por uma Matriz de Decis√£o, sendo composta de alternativas e crit√©rios ponderados de acordo com o tomador de decis√£o em quest√£o. M√©todos para solucionar problemas MCDM t√™m sido amplamente utilizados para selecionar a melhor alternativa entre um n√∫mero finito de alternativas.



-----------------------------------------------------------------------------------------------------------------------



**M√©todos AHP:**



O Analytic Hierarchy Process (AHP) √© um m√©todo para auxiliar √†s pessoas na tomada de decis√µes complexas. Mais do que determinar qual a decis√£o correta, o AHP ajuda as pessoas a escolher e a justificar a sua escolha. 



O AHP √© um m√©todo de tomada de decis√£o multicrit√©rio simples de usar e flex√≠vel. Permite a avalia√ß√£o de problemas complexos, baseados em crit√©rios conflitantes entre si. √â adequado para diferentes problem√°ticas, uma vez que a mesma dependa da intera√ß√£o humana para conduzir o processo de compara√ß√£o. Entre as vantagens de usar AHP como modelo de an√°lise decis√≥ria, √© a viabiliza√ß√£o de considerar diferentes n√≠veis de import√¢ncias entre as vari√°veis



Baseando-se nos modelos de Apoio Multicrit√©rio √† Decis√£o (AMD), o m√©todo AHP baseia-se em uma an√°lise hier√°rquica das vari√°veis. Neste modelo, um conjunto de alternativas √© avaliado segundo as prefer√™ncias do agente decisor quanto aos crit√©rios de influ√™ncia, buscando uma alternativa favor√°vel para o alcance eficiente do objetivo estrat√©gico.



Uma das principais caracter√≠sticas do m√©todo √© viabiliza√ß√£o de trabalhar com quest√µes subjetivas quanto as atribui√ß√µes de desempenho das vari√°veis do problema. Simplificando, na aplica√ß√£o do m√©todo AHP, n√£o √© necess√°rio definir um n√∫mero preciso quanto ao desempenho de uma alternativa em um dado crit√©rio, onde uma atribui√ß√£o verbal j√° satisfaz o modelo. As atribui√ß√µes do m√©todo AHP s√£o realizadas com base em uma escala proposta por Thomas Saaty



-----------------------------------------------------------------------------------------------------------------------

### $$Caracteristica~ do~ m√©todo ~AHP:$$



-----------------------------------------------------------------------------------------------------------------------



**Monot√¥nico de Lucro:** quanto maior, melhor





**Monot√¥nico de Custo:** quanto menor, melhor



-----------------------------------------------------------------------------------------------------------------------





**Transitividade:** Se A  B e B  C , ent√£o: A  C



-----------------------------------------------------------------------------------------------------------------------





**Teste de Consist√™ncia:** O AHP viabiliza calcular uma raz√£o de consist√™ncia (CR) mediante a compara√ß√£o do √çndice de Consist√™ncia (CI) das atribui√ß√µes do decisor quanto a um √≠ndice de consist√™ncia de uma matriz aleat√≥ria (RI)

```python
#Importar as bibliotecas
import numpy as np
import pandas as pd
```

# 1¬∞ Estrutura√ß√£o do problema

```python
# Criterios

criterios = ['custo','camera','armazenamento','bateria']
```

```python
#Alternativas

alt = ['xiaomi','samsung','iphone']
```

**Ap√≥s ter definido os crit√©rios e alternativas, vamos analisar os crit√©rios, viabilizando a obten√ß√£o de suas respectivas import√¢ncias**

## 2¬∞ Avalia√ß√£o Par a Par dos crit√©rios: Matriz de Pondera√ß√µes(ou Jugamentos)

```python
#Criar martiz de 1
matriz = pd.DataFrame(np.ones((4, 4),dtype=int),columns=criterios)
#Add nome nas linhas
matriz.index = criterios
matriz
```

**Criado a matriz com valor 1, pois o crit√©rio em diagonal est√° sendo comparado com ele mesmo**

## 3¬∞ Preencher a matriz com base na Escala Saaty

**Grau de Import√¢ncia:**


Custo VS Camera = 3

Custo VS armazenamento = 5

Custo VS bateria = 7

camera VS armazenamento = 3

camera VS bateria = 7

armazenamento VS bateria = 3


**Rec√≠proca:**

Camera VS Custo = 1/3

armazenamentoVS Custo  = 1/5

bateria VS Custo = 1/7

armazenamento VS camera = 1/3

bateria VS camera = 1/7

bateria VS armazenamento = 1/3


**O preenchimento dos valores de acordo com o crit√©rio √© de forma subjetiva**

```python
#Preencher
matriz['custo'] = [1,1/3,1/5,1/7]
matriz['camera'] = [3,1,1/3,1/7]
matriz['armazenamento'] = [5,3,1,1/3]
matriz['bateria'] = [7,7,3,1]

matriz_saaty = round(matriz,3)
matriz_saaty
```

**Observe o Axioma da Reciprocidade, Se custo VS camera = 3, ent√£o camera VS custo = 1/3**

**Obs: Diagonal com valor 1 se manteve**

## 4¬∞ Processo de normaliza√ß√£o (deixar os dados das colunas no formato de porcentagem)

```python
#Transposta para somar as colunas
matrizt = matriz_saaty.T

# Somar as colunas
matrizt['Soma_Total'] = matrizt.aggregate(np.sum,axis=1)

#Voltar a matriz original
matriz = matrizt.T
matriz
```

**Criado uma linha com o somat√≥rio total das colunas para o processo de normaliza√ß√£o**

```python
#Formula da normaliza√ß√£o (dado da Coluna / dado da Soma_Total)
matriz = matriz/matriz.iloc[4,:]
matriz = matriz.iloc[0:4,:]
matriz
```

**Matriz de pondera√ß√µes normalizados**

## 5¬∞ Criar Vetor de prioridade

```python
#Formula (Soma das linhas / numero de colunas)

matriz['Vetor_prioridade'] = matriz.aggregate(np.sum,axis=1)/len(matriz.columns)
matriz
```

**Vetor de prioridades s√£o os pesos de cada crit√©rio (em porcentagem), onde o somat√≥rio √© 100%**

## 6¬∞ Calcular a Raz√£o de Consist√™ncia

```python
# Calculo: matriz com base na Escala Saaty * Vetor de prioridades

consistencia = matriz_saaty * matriz['Vetor_prioridade']
consistencia
```

**Distribui√ß√£o da import√¢ncia de cada crit√©rio de acordo com a escala Saaty**

## 7¬∞ Soma dos pesos da Consist√™ncia

```python
## Calculo da soma dos pesos(Soma das linhas)

consistencia['soma_pesos'] = consistencia.sum(axis=1)
consistencia
```

**Vari√°vel soma_pesos √© importante para definir o lambda m√°ximo e calcular o resultado da C.I**

## 8¬∞ Obten√ß√£o do lambda m√°ximo

```python
# Calculo da obten√ß√£o do lambda m√°ximo

# Calculo 1: (Soma_pesos/Vetor_prioridade)
lambda_calculo1 = consistencia['soma_pesos']/matriz['Vetor_prioridade']
lambda_calculo1
```

```python
# Calculo 2: soma dos resultados
lambda_calculo2 = lambda_calculo1.sum()
lambda_calculo2
```

```python
#Calculo do lambda m√°ximo

#Calculo 3: soma dos resultados/ numero de crit√©rios
lambda_maximo = lambda_calculo2/len(lambda_calculo1)
lambda_maximo
```

**Valor do lambda m√°ximo**

## 9¬∞ Calcular o C.I

```python
# Formula do C.I:

#C.I = (Lambda m√°ximo - numero crit√©rios) / (numero crit√©rios -  1)

c_i = (lambda_maximo - len(lambda_calculo1)) / (len(lambda_calculo1) - 1)
c_i
```

**Valor do C.I**

## 10¬∞ Calcular o C.R

```python
# Formula do C.R:

# C.R = C.I / R.I
```

```python
#Matriz RI
from IPython.display import Image
Image(filename="RI.png")
```

**Veja que a matriz de ordem 4 √© usado o valor de 0.9 para calcular a C.R**

```python
#Calculo do C.R

c_r = c_i/0.9

#Converter para Porcentagem
round(c_r*100,2)
```

**Valor da raz√£o de consist√™ncia √© de 5.21%**

**Como o valor da raz√£o de consist√™ncia foi ‚â§ 10%, ent√£o o julgamento do decisor foi bom, pois ele foi consistente e preservou a transitividade matem√°tica**

### $$Conceito:$$

O M√©todo AHP-Gaussiano apresenta uma nova abordagem ao m√©todo original, baseada em uma an√°lise de sensibilidade proveniente do fator gaussiano

Com essa abordagem √© poss√≠vel obter os pesos dos crit√©rios mediante as entradas quantitativas das alternativas em cada crit√©rio em an√°lise, ou seja, a partir dos dados da pr√≥pria matriz de decis√£o.

Um dos pontos de destaque do novo modelo relaciona-se a redu√ß√£o do esfor√ßo cognitivo do decisor, a partir do momento em que n√£o √© mais necess√°rio a avalia√ß√£o par a par entre os crit√©rios para obten√ß√£o de seus respectivos pesos

Entretanto, vale ressaltar, que a viabilidade do modelo s√≥ √© satisfeita em cen√°rios em que as alternativas possuam entradas cardinais nos crit√©rios em an√°lise

## Etapa 1 ‚Äì Determina√ß√£o da Matriz de Decis√£o

```python
# Dados originais

#Criar martiz de 1
dados = pd.DataFrame(np.ones((3, 4),dtype=int),columns=criterios)
#Add nome nas linhas
dados.index = alt

#Alterar os valores com os dados originais
dados['custo'] = [1500, 1800 , 5000]
dados['camera'] = [12,12,20]
dados['armazenamento'] = [64,128,128]
dados['bateria'] = [24,18,10]

#Dados originais
dados
```

**Dados originais**

```python
# Normalizar os dados
dados_norm = dados.copy()

# Monot√¥nico de Custo: Inverter o valor para normalizar = 1/valor
dados_norm['custo'] = 1/dados_norm['custo']
# Somatorio do custo
dados_norm['custo'].sum()
#Indice de preferencia local Monot√¥nico de Custo
dados_norm['custo'] = dados_norm['custo']/dados_norm['custo'].sum()
#Indice de preferencia local Monot√¥nico de Lucro
dados_norm['camera'] = dados_norm['camera']/dados_norm['camera'].sum()
dados_norm['armazenamento'] = dados_norm['armazenamento']/dados_norm['armazenamento'].sum()
dados_norm['bateria'] = dados_norm['bateria']/dados_norm['bateria'].sum()
dados_norm 
```

**Dados normalizados no formato de porcentagem**

## Etapa 2 ‚Äì C√°lculo da m√©dia das alternativas em cada crit√©rio

```python
#M√©dia da base de dados = Somat√≥rio das colunas / n√∫mero de crit√©rios

media = dados_norm.sum(axis=0)/3
media
```

**Valor da m√©dia de cada coluna**

## Etapa 3 ‚Äì C√°lculo do desvio padr√£o dos crit√©rios com base na amostra das alternativas


```python
#Desvio padr√£o da base de dados original
desv = dados_norm.std()
desv
```

**desvio padr√£o de cada criterio**

## Etapa 4 ‚Äì C√°lculo do fator gaussiano para cada crit√©rio

```python
#Fator gaussiano
f_gaussiano = desv/media
f_gaussiano
```

**Fator gaussiano de cada crit√©rio**

```python
#Somatorio do Fator gaussiano
f_gaussiano.sum()
```

**Note que o fator gaussiano n√£o da 1, portanto teremos que normalizar os dados**

```python
#Transformar em porcentagem o fator gaussiano
f_gau_porcent = f_gaussiano / f_gaussiano.sum()
print(f_gau_porcent)
print(f'\nSomat√≥ria: {f_gau_porcent.sum()}')
```

**Note que agora a somat√≥ria da 1**

## Etapa 5 ‚Äì Pondera√ß√£o da matriz de decis√£o 

```python
#Pondera√ß√µes dos dados
pond_matriz = dados_norm * f_gau_porcent
pond_matriz
```

**Definida a import√¢ncia de cada crit√©rio**

## Etapa 6 ‚Äì Obten√ß√£o do Ranking

```python
#Ranking
pond_matriz['Soma_Ponderacoes'] = pond_matriz.sum(axis=1)
pond_matriz
```

**Com base no valor da pondera√ß√£o de cada crit√©rio, podemos definir a ordem do melhor produto sendo: 1¬∞ Xiaomi, 2¬∞ Samsung e 3¬∞ Iphone**

# Compara√ß√£o dos 2 M√©todos

**Observe que os 2 m√©todos tiveram a mesma ordem de import√¢ncia dos produtos, com uma pequena varia√ß√£o no valor final das pondera√ß√µes**
