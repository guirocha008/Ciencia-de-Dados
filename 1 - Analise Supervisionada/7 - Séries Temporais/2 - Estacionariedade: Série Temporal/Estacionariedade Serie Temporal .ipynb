# $$Estacionariedade:Série ~Temporal $$

### $$Conceito:$$


Um processo estacionário tem a propriedade de que a média, variância e estrutura de autocorrelação não mudam no decorrer do tempo. Os métodos de modelagem estatística assumem ou exigem que as séries temporais sejam estacionárias.


As observações em uma série temporal estacionária não dependem do tempo.


**Há duas razões principais por trás da não estacionariedade de uma série temporal:** 


- Tendência e Sazonalidade


**Tipos de Estacionariedade:**


**Estacionaridade Estrita**: Na estacionariedade forte ou estrita a forma de distribuição conjunta do processo permanece sem variação mediante uma transação temporal 


**Estacionaridade de Tendência**: Uma série que não possui raiz unitária, mas exibe uma tendência, O teste KPSS classifica uma série como estacionária na ausência de raiz da unidade. Isso significa que a série pode ser estacionária estrita ou estacionária de tendência.


**Estacionaridade Diferencial**: Uma série temporal que pode ser tornada estritamente estacionária pela diferenciação é do tipo estacionária diferencial. O teste ADF indica a quantidade de diferenciação necessária para tornar a série estacionaria

-----------------------------------------------------------------------------------------------------------------------

### $$Decomposição ~da~ Série~ Temporal:$$
    
Conseguir avaliar os diferentes elementos que compõem a serie temporal (Tendencia, Ciclo, Sazonalidade e Resíduos)


Nesta abordagem, a tendência e a sazonalidade são modeladas separadamente e a parte restante da série é retornada.


**Componentes:**

    
**Tendencia:** É o padrão de longo prazo de uma série temporal. Uma tendência pode ser positiva ou negativa, dependendo se a série temporal exibe um padrão crescente de longo prazo ou um padrão decrescente de longo prazo, podendo ser linear ou não linear (curvilínea). Se  uma  série  temporal  não  mostrar  um  padrão  crescente  ou  decrescente,  a  série  será estacionária
    

**Ciclo:** Caracterizado pela oscilação de longo prazo de subida e descida nas séries (de forma suave e repetida) ao longo do componente da tendência, A variação cíclica é de natureza periódica e se repete como um ciclo comercial, que possui quatro fases: Pico, Recessão, Depressão e Expansão

    
    
**Sazonalidade:** São flutuações regulares por um período menor que um ano (Dia, Mês, Trimestre, Ano) 
    
    
**Resíduos:** Efeitos que não foram incorporados pela série de tempo, ou seja, resíduo


    
**A decomposição da série pode ter modelos Aditivo ou Multiplicativo**


- **Aditivo:** Série formada pela soma dos componentes (Tendencia, Ciclo, Sazonalidade e Resíduos), O modelo aditivo geralmente é considerado mais adequado para dados em que as flutuações sazonais permanecem aproximadamente do mesmo tamanho com o tempo.
    
    
- **Multiplicativo:** Série formada pela multiplicação dos componentes (Tendencia, Ciclo, Sazonalidade e Resíduos), O modelo multiplicativo é normalmente aplicado a dados em que o tamanho dos efeitos sazonais aumenta ao longo do tempo.

-----------------------------------------------------------------------------------------------------------------------

### $$Séries: Univariada ~x ~Multivariada:$$


As séries temporais **univariadas** dependem de uma única variável, enquanto o conjunto de dados de séries temporais **multivariadas** contém diversas variáveis dependentes, que são levadas em consideração na previsão de valores futuros


Exemplo: Para séries temporais univariadas, usamos apenas dados históricos de temperatura para prever valores futuros. Para séries temporais multivariadas, combinamos 5 séries temporais (temperatura, nebulosidade, ponto de orvalho, umidade e vento) 

-----------------------------------------------------------------------------------------------------------------------

### $$Classificação:Série~Temporal:$$


**Estacionaria:** Os dados variam em torno da média ao longo do tempo, media igual a zero e variância constante, sem efeito sazonal, propriedades estatísticas não mudam ao longo do tempo, as teorias relacionadas às séries estacionárias são mais maduras e mais fáceis de implementar em comparação às séries não estacionárias.


**Não Estacionaria:** Segue uma tendência e os dados não retorna a media, são imprevisíveis e não podem ser modelados ou previstos


**Determinística:** Quando pode ser descrita por uma função matemática para estabelecer exatamente os valores futuros da série.


**Estocástica:** Quando os valores futuros da série somente podem ser estabelecidos em termos probabilísticos, pois o modelo compõe-se também de um termo aleatório.


-----------------------------------------------------------------------------------------------------------------------


### $$Autocorrelação:$$

É a **correlação** de uma série temporal com uma **cópia atrasada** (lag, atraso ou defasagem) de si mesma.


Observação num determinado instante está relacionado às **observações passadas**


A ocorrência de **não estacionariedade** é denotada pela **lenta queda da ACF** nos primeiros lags da série.


As observações podem estar autocorrelacionadas em diversas ordens. A **autocorrelação de primeira ordem** caracteriza séries onde uma observação está correlacionada com a **observação imediatamente anterior** (janeiro e dezembro, por exemplo). A autocorrelação de **segunda ordem** caracteriza séries temporais onde uma observação está correlacionada com as **observações a 2 unidades de tempo no passado** (janeiro e novembro, por exemplo)



**Ruido Branco** - Não apresenta Autocorrelação


-----------------------------------------------------------------------------------------------------------------------


### $$Diferenciação:$$


**Dados não estacionários** devem ser transformados em dados estacionários através da diferenciação, para que  sejam consistentes e confiáveis


Às vezes, a operação de **diferenciação  deve  ser  repetida**  um  certo  número  de  vezes  até  que  a função resultante fique estacionária


**Tempo de atraso (defasagem)** na série para calcular as correlações, as defasagens são criadas após o deslocamento da série e, em seguida, compare a série defasada com a série original sem defasagem

-----------------------------------------------------------------------------------------------------------------------

### $$Smoothing (alisamento~ ou ~suavização):$$

Usamos **suavização** quando queremos remover as flutuações nos dados e focamos apenas em preservar as tendências de longo prazo


**Técnicas de Smoothing:** Média Móvel Simples, Média Móvel Ponderada Exponencial


**Uma média móvel** requer que você especifique um tamanho de janela chamado largura da janela. Isso define o número de observações brutas usadas para calcular o valor da média móvel

-----------------------------------------------------------------------------------------------------------------------

### $$LOG ~Natural:$$

Transforma os dados em uma **distribuição normal** ou próximo dela


Quando **não conseguimos identificar uma tendencia nos dados**, transformamos toda base em LOG


**Reduz o efeito dos ruídos**

## Explorando as Séries Temporais

```python
#Dataset 
from statsmodels.tsa.seasonal import seasonal_decompose
df = pd.read_csv('diabetic.csv',sep=';',decimal=',')
df['time'] = pd.date_range('1991-07-01',freq='M',periods=204)
df = df.set_index('time')
df.info()
```

###  Decomposição da Série Temporal

```python
# Decompondo a série temporal
decomposition = seasonal_decompose(df)

# Extraindo cada componente da série
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid


# Plot
plt.subplot(411)
plt.plot(df['value'], label = 'Série Original')
plt.legend(loc = 'best')
plt.subplot(412)
plt.plot(trend, label = 'Tendência')
plt.legend(loc = 'best')
plt.subplot(413)
plt.plot(seasonal,label = 'Sazonalidade')
plt.legend(loc = 'best')
plt.subplot(414)
plt.plot(residual, label = 'Resíduos')
plt.legend(loc = 'best')
plt.tight_layout()
```

**Observe que a decomposição dos dados possui 4 gráficos: Original, Tendência, Sazonalidade e Resíduos**

## Extraindo os componentes: Tendência, Sazonalidade e Resíduo

```python
# Extraindo cada componente da série
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid


# Observações = (Seasonal * Trend * Resid)
df_reconstructed = pd.concat([trend, seasonal, residual, decomposition.observed], axis = 1)
df_reconstructed.columns = ['Trend','Seasonal', 'Resid', 'Valores_Observados']
df_reconstructed.head(10)
```

# Estacionariedade

**Podemos verificar a estacionariedade de uma série temporal de algumas formas:**


- **1° - Plot das Rolling Statistics (Estatísticas Móveis)**: Podemos criar um plot da média móvel e variância móvel (ou desvio padrão, que é a raiz quadrada da variância) e ver se variam com o tempo.


- **2° - Autocorrelação com Gráficos ACF e PACF**: Se a série temporal for estacionária, os gráficos ACF / PACF, mostrarão um padrão específico


- **3° - Teste Dickey-Fuller**: Testes estatísticos para verificar a estacionariedade, a hipótese nula é que a série temporal não é estacionária.

### 1°- Plot das Rolling Statistics (Estatísticas Móveis)

- Primeiro, vamos checar se as estatísticas móveis são ou não constantes ao longo da série temporal.

```python
# Prepara os dados
x = df.index.values # Tempo
y = df # Série

# Determinando estatísticas móveis (Série anual)
rolmean = df.rolling(window = 12).mean()
rolstd = df.rolling(window = 12).std()

# Plot das estatísticas móveis
plt.figure(figsize=(10, 6))
orig = plt.plot(y, color = 'blue', label = 'Original')
mean = plt.plot(rolmean, color = 'red', label = 'Média Móvel')
std = plt.plot(rolstd, color = 'black', label = 'Desvio Padrão')
plt.legend(loc = 'best')
plt.title('Estatísticas Móveis - Média e Desvio Padrão')
plt.show()
```

**Média móvel e desvio padrão não são constantes em relação ao tempo (tendência crescente), série não estacionaria**

### 2° - Plots ACF e PACF

- No gráfico, o eixo vertical indica a autocorrelação e o eixo horizontal o número de defasagem(Lag)

```python
#Auto Correlação
plt.figure(figsize=(12, 5))
plot_acf(df.Passengers,plt.gca(), lags=30)
plt.show()
```

**Como é possível ver na imagem, temos diversos valores ACF acima do limite da área sombreada em azul. Nesses casos, a autocorrelação é diferente de zero, indicando que a série não é aleatória**

```python
# Plot do gráfico PACF
plt.figure(figsize=(12, 5))
plot_pacf(df.Passengers, ax = plt.gca(), lags = 30)
plt.show()
```

**O PACF é uma visualização parcial do ACF, podemos observar vários pontos fora do intervalo de confiança de 95%, portanto há indícios de uma série não estacionaria**

### 3° - Teste de Dickey-Fuller

- Hipótese nula: a série temporal não é estacionária.


- Rejeitar a hipótese nula (ou seja, um valor-p abaixo de 0.05) indicará estacionaridade.


```python
# Teste Dickey-Fuller

from statsmodels.tsa.stattools import adfuller



# Print
print('\nResultado do Teste Dickey-Fuller:\n')

# Teste
dfteste = adfuller(df.Passengers, autolag = 'AIC')

# Formatando a saída
dfsaida = pd.Series(dfteste[0:4], index = ['Estatística do Teste',
                                           'Valor-p',
                                           'Número de Lags Consideradas',
                                           'Número de Observações Usadas'])



# Print
print (dfsaida)
```

**O Valor-p é maior que 0.05. Não temos evidências para rejeitar a hipótese nula de que a série não é estacionária**

**OBS: Se a série não for estacionária, teremos que aplicar transformações antes da modelagem**

## Transformações em Séries Temporais - Log, Raiz Quadrada e Box-Cox

- Esta é uma técnica de transformação de dados útil usada para estabilizar a variância, tornar os dados mais semelhantes à distribuição normal


- **logarítmica** = É usado para transformar séries que tem valores com crescimentos exponenciais em séries com crescimentos mais lineares


- **Raiz Quadrada** = linearizar os dados calculando sua raiz quadrada.


- **Box-Cox** = Deixar os dados mais normalizados e consequentemente estabiliza a variância durante os dados




## 1° - Transformação de Log


- Séries temporais com uma distribuição exponencial podem ser linearizadas usando o logaritmo dos valores


- É importante observar que esta operação assume que os valores são positivos e diferentes de zero

```python
#Dataset 
df = pd.read_csv('AirPassengers1.csv',sep=',')
df['Month'] = pd.to_datetime(df.Month)
df = df.set_index('Month')


# Vamos aplicar uma transformação de log usando np.log() e gravamos o resultado em uma nova coluna do nosso dataset
df['Passageiros_log'] = np.log(df.Passengers)

# Plot da séries
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].plot(df['Passengers'], color = "red");
axes[0].legend(['Antes da Transformação'])
axes[1].plot(df['Passageiros_log'], color = "green");
axes[1].legend(['Após a Transformação']);
```

**Note que após a transformação, a série que parecia um modelo multiplicativo passa a ser aditivo, e aparentemente mais linear, porém ainda não estacionaria, pois, está com tendência**

```python
# Scatter das séries
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].scatter(df.index,df['Passengers'], color = "red");
axes[0].legend(['Antes da Transformação'])
axes[1].scatter(df.index,df['Passageiros_log'], color = "green");
axes[1].legend(['Após a Transformação']);
```

**Note que os dados ficam mais lineares após a transformação de LOG**

```python
# Histograma da séries 
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].hist(df['Passengers']); #dados original
axes[0].legend(['Antes da Transformação'])
axes[1].hist(df['Passageiros_log']);#log natural
axes[1].legend(['Após a Transformação']);
```

**O histograma mostra uma distribuição de observações do tipo gaussiana após a transformação**

```python
# Média da série antes e depois 
a= df['Passengers'].mean()
b = df['Passageiros_log'].mean()

print(f'Média Original: {round(a,3)}')
print(f'Média após a transformação de Log: {round(b,3)}')
```

```python
# Desvio Padrão da série antes e depois
a = df['Passengers'].std()
b = df['Passageiros_log'].std()

print(f'Desvio Padrão Original: {round(a,3)}')
print(f'Desvio Padrão após a transformação de Log: {round(b,3)}')
```

**Note que a transformação do log por si só não torna a série estacionaria, é preciso aplicar a diferenciação, para eliminar a tendência e a sazonalidade**

## 2° - Transformação de Raiz Quadrada


- Uma série temporal que tem uma tendência quadrática de crescimento pode ser linearizada calculando sua raiz quadrada.



```python
# Transformação de Raiz Quadrada com np.sqrt()
dados_sqrt = np.sqrt(df['Passengers'])

# Gráfico de Linha
# Plot da séries
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].plot(df['Passengers'], color = "blue");
axes[1].plot(dados_sqrt, color = "magenta");
axes[0].legend(['Antes da Transformação'])
axes[1].legend(['Após a Transformação']);
```

**Podemos ver que a tendência foi reduzida, mas não foi removida.**

```python
# Scatter das séries
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].scatter(df.index,df['Passengers'], color = "red");
axes[1].scatter(df.index,dados_sqrt, color = "green");
axes[0].legend(['Antes da Transformação'])
axes[1].legend(['Após a Transformação']);
```

**Extraindo a raiz quadrada dos dados, altera sua linearidade**

```python
# Média da série antes e depois da diferenciação
a = df['Passengers'].values.mean()
b = dados_sqrt.mean()

print(f'Média:\n')
print(f'Média Original:  {round(a,3)}')
print(f'Média após extrair a raiz quadrada: {round(b,3)}')

print(f'\nDesvio Padrão:\n')

a = df['Passengers'].values.std()
b = dados_sqrt.std()

print(f'Dev Padrão Original:  {round(a,3)}')
print(f'Dev Padrão após extrair a raiz quadrada:: {round(b,3)}')
```

**Observe que tanto a média quanto o desvio padrão teve uma queda drástica em seus valores**

```python
# Histograma da séries 
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].hist(df['Passengers']); #dados original
axes[1].hist(dados_sqrt, color = "magenta")
axes[0].legend(['Antes da Transformação'])
axes[1].legend(['Após a Transformação']);
plt.show()
```

**Extraindo a Raiz quadrada, não transforma os dados em uma distribuição gaussiana**

## 3° - Transformação Box-Cox

- A transformação Box-Cox é um método de transformação de dados configurável que suporta a raiz quadrada e a transformação de log, bem como um conjunto de transformações relacionadas


- A série resultante pode ser mais linear e a distribuição resultante mais gaussiana ou uniforme, dependendo do processo que a gerou


**Abaixo estão alguns valores comuns para lambda**

- lambda = -1. é uma transformação recíproca.
- lambda = -0,5 é uma transformação de raiz quadrada recíproca.
- lambda = 0,0 é uma transformação de log.
- lambda = 0,5 é uma transformação de raiz quadrada.
- lambda = 1.0 não é transformação.

```python
# Aplicando Transformação Box-Cox, defindo o lambda sendo zero
df['Passageiros_box'] = boxcox(df['Passengers'], lmbda = 0.0)

# Plot da série transformada
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].plot(df['Passengers'], color = "blue");
axes[1].plot(df['Passageiros_box'], color = "magenta");
axes[0].legend(['Antes da Transformação'])
axes[1].legend(['Lambda = 0']);
```

**Note que após a transformação, a série que parecia um modelo multiplicativo passa a ser aditivo, e aparentemente mais linear, porém ainda não estacionaria, pois, está com tendência**

```python
# Scatter das séries
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].scatter(df.index,df['Passengers'], color = "red");
axes[1].scatter(df.index,df['Passageiros_box'], color = "green");
axes[0].legend(['Antes da Transformação'])
axes[1].legend(['Lambda = 0']);
```

**Note que as dados ficam lineares após a transformação**

```python
# Histograma da séries 
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(15,5))
axes[0].hist(df['Passengers']); #dados original
axes[1].hist(df['Passageiros_box'], color = "magenta")
axes[0].legend(['Antes da Transformação'])
axes[1].legend(['Lambda = 0']);
plt.show()
```

**O histograma mostra uma distribuição de observações do tipo gaussiana após a transformação**

### OBS: Podemos definir o parâmetro lambda como None (o padrão) e deixar a função encontrar um valor ajustado estatisticamente.

# Smoothing (Suavização ou Alisamento)

### Média Móvel Simples

- Calculamos a média de valores 'k' consecutivos, dependendo da frequência da série temporal


- Uma desvantagem dessa abordagem específica é que o período de tempo deve ser estritamente definido

```python
# Carrega os dados
# Observe que estamos carregando os dados e já definindo a data como índice com o parâmetro index_col = 0
dados_serie = pd.read_csv('https://raw.githubusercontent.com/dsacademybr/Datasets/master/dataset1.csv', 
                          header = None, 
                          index_col = 0,
                          names = ['Serie_Original_Passageiros'])

#Criar uma copia
df_serie = dados_serie.copy()

#Período da séries para calcular a média movel
dados_serie
```

**Note que o período para usar a média móvel é mensal**

```python
# Agora definimos uma janela de 12 meses da série temporal para calcular a média móvel
dados_serie['Serie_Suavizada_Passageiros'] = dados_serie.rolling(window = 12).mean()

# Imprime o dataset 
dados_serie.head(13)
```

**Observe que usando uma média móvel de 12 meses, os primeiros 12 meses do dataset original são perdidos para o cálculo da média**

```python
#Plot da série original e a série suavizada com a média móvel
plt.figure(figsize=(17, 5))
plt.plot(dados_serie['Serie_Original_Passageiros'], color = 'blue')
plt.plot(dados_serie['Serie_Suavizada_Passageiros'], color = 'red')
plt.legend(['Serie_Original','Serie_Suavizada'])
plt.show()
```

**Série original e série suavizada com a média móvel**

```python
# Calculando a diferença entre a série original e a série suavizada
plt.figure(figsize=(17, 5))
diff_original_suavizada = dados_serie['Serie_Original_Passageiros'] - dados_serie['Serie_Suavizada_Passageiros']
plt.plot(diff_original_suavizada)
plt.show()
```

**Removendo a tendência da série original, aplicando a diferença entre a série original e a série suavizada**

```python
#Teste de estacionariedade

# Remover os valores NA para testar a estacionaridade
diff_original_suavizada.dropna(inplace = True)

from statsmodels.tsa.stattools import adfuller

# Teste Dickey-Fuller

# Print
print('\nResultado do Teste Dickey-Fuller:\n')

# Teste
dfteste = adfuller(diff_original_suavizada, autolag = 'AIC')

# Formatando a saída
dfsaida = pd.Series(dfteste[0:4], index = ['Estatística do Teste',
                                           'Valor-p',
                                           'Número de Lags Consideradas',
                                           'Número de Observações Usadas'])



# Print
print (dfsaida)


```

**O valor-p é menor que 0.05, portanto, temos evidências para rejeitar a hipótese nula, série provavelmente é estacionária**

```python
# Média da série antes e depois da diferenciação
a = dados_serie.mean()
b = diff_original_suavizada.mean()

print(f'Média:\n')
print(f'Média Original:  {round(a[0],3)}')
print(f'Média após o Smoothing: {round(b,3)}')

print(f'\nDesvio Padrão:\n')

a = dados_serie.std()
b = diff_original_suavizada.std()

print(f'Dev Padrão Original:  {round(a[0],3)}')
print(f'Dev Padrão após o Smoothing: {round(b,3)}')
```

**Mesmo a média não sendo zero e o desvio padrão não sendo constante, após a aplicação do Smoothing a série se tornou estacionaria**

**OBS: Com a suavização podemos transformar a  série em estacionária. Mas essas técnicas simples de suavização não funcionam em todos os casos, principalmente nos de alta sazonalidade. E para esses casos precisamos da Diferenciação**

# Diferenciação(Lag)


- O processo de diferenciação pode ser repetido mais de uma vez até que toda a dependência temporal seja removida.


- Remove a tendência e a sazonalidade da série, para ser estacionária


- 1° diferenciação normalmente elimina a tendência da série e a 2° diferenciação elimina a sazonalidade

```python
# Função para testar a estacionaridade
def testa_estacionaridade(serie):
    
    # Calcula estatísticas móveis
    rolmean = serie.rolling(window = 12).mean()
    rolstd = serie.rolling(window = 12).std()

    # Plot das estatísticas móveis
    plt.figure(figsize=(15, 6))
    orig = plt.plot(serie, color = 'blue', label = 'Original')
    mean = plt.plot(rolmean, color = 'red', label = 'Média Móvel')
    std = plt.plot(rolstd, color = 'black', label = 'Desvio Padrão')
    
    plt.legend(loc = 'best')
    plt.title('Estatísticas Móveis - Média e Desvio Padrão')
    plt.show()
    
    # Teste Dickey-Fuller:
    # Print
    print('\nResultado do Teste Dickey-Fuller:\n')

    # Teste
    dfteste = adfuller(serie, autolag = 'AIC')

    # Formatando a saída
    dfsaida = pd.Series(dfteste[0:4], index = ['Estatística do Teste',
                                               'Valor-p',
                                               'Número de Lags Consideradas',
                                               'Número de Observações Usadas'])

    # Loop por cada item da saída do teste
    for key, value in dfteste[4].items():
        dfsaida['Valor Crítico (%s)'%key] = value

    # Print
    print (dfsaida)
    
    # Testa o valor-p
    print ('\nConclusão:')
    if dfsaida[1] > 0.05:
        print('\nO valor-p é maior que 0.05 e, portanto, não temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente não é estacionária.')
    else:
        print('\nO valor-p é menor que 0.05 e, portanto, temos evidências para rejeitar a hipótese nula.')
        print('Essa série provavelmente é estacionária.')
```

```python
# Carrega os dados

# Observe que estamos carregando os dados e já definindo a data como índice com o parâmetro index_col = 0
dados_serie = pd.read_csv('https://raw.githubusercontent.com/dsacademybr/Datasets/master/dataset1.csv', 
                          header = None)

# Ajusta o nome das colunas
dados_serie.columns = ['data','passageiros']

# Converte a data no formato apropriado
dados_serie['data'] = pd.to_datetime(dados_serie['data'], format = '%Y-%m')


# Define a data como índice da série
dados_serie = dados_serie.set_index('data')
dados_serie.head()

```

```python
#Verificar tipo de dados
dados_serie.info()
```

```python
# Testar a estacionaridade da série antes de qualquer transformação
testa_estacionaridade(dados_serie)
```

```python
# Uma Diferenciação Para Remover Tendência 
dados_serie_diff1 = dados_serie.passageiros - dados_serie.passageiros.shift(1)

#Plot
plt.figure(figsize=(15, 6))
plt.plot(dados_serie_diff1);
```

```python
# Média da série antes e depois da diferenciação
a = dados_serie.values.mean()
b = dados_serie_diff1.mean()

print(f'Média:\n')
print(f'Média Original:  {round(a,3)}')
print(f'Média após a Diff: {round(b,3)}')

print(f'\nDesvio Padrão:\n')

a = dados_serie.values.std()
b = dados_serie_diff1.std()

print(f'Dev Padrão Original:  {round(a,3)}')
print(f'Dev Padrão após a Diff: {round(b,3)}')
```

**Note que a média apesar da queda ainda não está próxima de 0 e o desvio padrão está com valor alto**

```python
#Teste de estacionariedade com 1 diferenciação
testa_estacionaridade(dados_serie_diff1.dropna())
```

**Apesar do p_valor ser rejeitado, ele está bem proximo do valor de 0.05**

```python
#Aplicar 2 Diferenciação nos dados
dados_serie_diff2 = dados_serie.passageiros - dados_serie.passageiros.shift(2)

#Plot
plt.figure(figsize=(15, 6))
plt.plot(dados_serie_diff2);
```

**Série com 2 diferenciações fica estacionaria**

# Conclusão:

**Para prever os dados de uma série temporal ela precisar ser estacionaria, média constante durante o tempo e não deve existir tendência de alta ou baixa, para isso acontecer usamos alguns métodos:**

**Verificar se uma série é estacionaria ou não, através dos testes de estacionariedade: Estatísticas Móveis, Autocorrelação com Gráficos ACF e PACF e Teste Dickey-Fuller, caso a série seja estacionaria, pode partir para modelagem preditiva, caso contrário, a série precisa ser Pré-processada:**

**Série não estacionaria: verificar a distribuição dos dados, caso os dados não estejam com distribuição normal aplicar transformação de: Log, Box-Cox e Raiz quadrada**

**Após a distribuição da série ser ou ficar próxima de uma gaussiana, precisa tornar a série estacionaria, aplicando algumas técnicas como: Decomposição, Smoothings, Diferenciação, assim conseguimos deixar a série sem tendência e sazonalidade, pronta para modelagem preditiva**
