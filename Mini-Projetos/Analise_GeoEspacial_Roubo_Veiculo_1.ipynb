# $$An√°lise~GeoEspacial: Roubo~ de ~ve√≠culo$$

### $$Conceito:$$

- **Primeira lei da geografia** "Todas as coisas est√£o relacionadas com todas as outras, mas coisas pr√≥ximas est√£o mais relacionadas do que coisas distantes" (Tobler, 1970).


- **Dados espaciais** indicam o quanto varia dado fen√¥meno, preocupando-se com onde (lugar) ocorre tal varia√ß√£o; j√° os **dados n√£o espaciais** indicam o quanto varia determinado fen√¥meno, sem preocupa√ß√£o acerca do lugar em que ocorre essa varia√ß√£o


### $$Latitude~e~Longitude:$$

Dist√¢ncias que permitem a localiza√ß√£o de qualquer ponto na superf√≠cie do planeta Terra. Elas formam o sistema de coordenadas geogr√°ficas, e correspondem √† dist√¢ncia em graus medida com base na Linha do Equador, no caso da **latitude**, e no Meridiano de Greenwich, no caso das **longitudes**


- **Latitude** √© a dist√¢ncia em graus de um ponto na superf√≠cie terrestre at√© a Linha do Equador. Ela vai de 0 a 90¬∫ tanto no Hemisf√©rio Norte quanto no Hemisf√©rio Sul

- **Longitude** √© a dist√¢ncia em graus de um ponto na superf√≠cie terrestre at√© o Meridiano de Greenwich. Distribui-se entre o intervalo de 0 a 180¬∫ no Hemisf√©rio Leste e no Oeste


### $$Coordinate~Reference~System(CRS):$$

Uma **CRS** indica coordenadas na superf√≠cie terrestre (e.g. longitude e latitude), mas aten√ß√£o na escolha do tipo, pois h√° coordenadas que consideram medidas geod√©sicas, outras consideram medidas euclidianas; outras levam em considera√ß√£o o centro de massa do planeta; outras assumem pontos arbitr√°rios de interesse



### $$Classes:$$


**Shapefiles:**

- Os **shapefiles** s√£o arquivos que cont√™m informa√ß√µes de determinada geografia, incluindo a sua localiza√ß√£o e o seu formato


- Os **shapefiles** correspondem a conjuntos de arquivos que possibilitam a exist√™ncia de um mapa atrelado a uma base de dados. Esses arquivos, no m√≠nimo, possuem extens√µes ***.shp, *.shx, *.dbf e *.prj**
    - ***.shp:** arquivo que cont√©m a geometria, isto √©, os pol√≠gonos que compor√£o o mapa
    - ***.dbf:** arquivo que cont√©m a base de dados
    - ***.shx:** arquivo que relaciona os arquivos *.shp e *.dbf
    - ***.prj:** arquivo que descreve qual o sistema de proje√ß√£o geogr√°fica o mapa utiliza
    
    
**Simple Feature:**

- **Simple feature** referem-se, em regra, a data frames que possuam vetores com informa√ß√µes de refer√™ncia geogr√°fica



**Spatial Points:**


- S√£o objetos semelhantes aos da classe **Simple Feature**, podendo ou n√£o, possuir uma base de dados atrelada


**Raster:**

- Objetos **raster** s√£o aqueles que cont√™m imagens com descri√ß√µes de cada um dos pixels que os comp√µem. Essa esp√©cie de recurso traz, no lugar de pol√≠gonos ou pontos, uma imagem (formato *.tiff, *.jpeg, *.bmp, etc.) georreferenciada


### $$Algumas~t√©cnicas~de~An√°lise~Espacial:$$

**Buffer Analysis:**

- **Buffering** √© uma t√©cnica para se medir dist√¢ncias para fora, a partir do centro de um dado ponto geogr√°fico


**Buffer Union:**

- **Buffer Union** √© uma t√©cnica que combina os pontos de encontro dos outputs da t√©cnica de buffering


**Kernel Densities:**

- A t√©cnica de **kernel densities** mensura a densidade da presen√ßa de pontos de interesse em determinada √°rea geogr√°fica


### $$Massificando~o~conceito~de~Datum:$$

- No campo da An√°lise Espacial, podemos entender o conceito de **datum** como um conjunto de informa√ß√µes que engloba um sistema de pontos de refer√™ncia na superf√≠cie terrestre que se conecta (ou deve se conectar) ao modelo da forma da Terra (planar, elipsoide, etc.) para que se possa definir um sistema de coordenadas geogr√°ficas


- O **datum** mais comumente utilizado diz respeito ao World Geodetic System 1984 (WSG 84), tamb√©m conhecido como WGS 1984, EPSG:4326


**WSG 84:**

- o **WSG 84** configura-se numa esp√©cie de norma utilizada pela Cartografia que possui um sistema de coordenadas terrestres (latitude e longitude), assumindo a Terra como uma elipsoide ‚Äì para efeitos de sua superf√≠cie e altitudes ‚Äìe uma equipotential gravitational surface (geoide) para seu n√≠vel do mar


- **OBS:** H√° outras normas que podem ser utilizadas, a depender de como se assume o formato da Terra e de como se mensuram suas coordenadas, bem como a partir de qual ponto de refer√™ncia a essas coordenadas se orientam (e.g. EPSG:3857, EPSG:7789, Corrego Alegre, etc.)


### $$An√°lise~Explorat√≥ria~de~Dados~Espaciais:$$


-  A **Exploratory Spatial Data Analysis (ESDA)** direciona-se a facilitar a identifica√ß√£o e a visualiza√ß√£o de padr√µes espaciais de um dado fen√¥meno


- O primeiro passo para efetuarmos uma **ESDA** √© o estabelecimento das vizinhan√ßas entre as localidades estudadas para que, ent√£o possamos, por exemplo, verificar as **autocorrela√ß√µes espaciais**, apontar algumas **heterogeneidades** e, inclusive, detectar eventuais **outliers**

**Autocorrela√ß√£o espacial:**

- Mede a **correla√ß√£o** da vari√°vel com ela mesma, mas considerando **diferentes dist√¢ncias** entre as observa√ß√µes


- Uma t√©cnica importante para a **an√°lise explorat√≥ria dos dados (ESDA)** e tamb√©m para utilizar no c√°lculo do √≠ndice de autocorrela√ß√£o espacial √© a **matriz de vizinhan√ßa**


- **Passo 1:** Escolher um **crit√©rio de vizinhan√ßa**
    - √â de suma import√¢ncia convencionar as formas de vizinhan√ßa nas **matrizes de vizinhan√ßa** que consideram a **contiguidade (matriz bin√°ria)** ou  **dist√¢ncias**(KNN). Sendo assim, a contiguidade segue as regras dos movimentos de pe√ßas que existem no jogo de xadrez, tais como a rainha (Queen), a torre (Rook) e o bispo (Bishop), onde os elementos bin√°rios apontam para exist√™ncia ou aus√™ncia de vizinhan√ßa entre um par qualquer de entidades
    - **OBS:** Os m√©todos mais comuns envolvem a medida de **contiguidade** entre pol√≠gonos, ou seja, a conex√£o/interse√ß√£o entre arestas de dois pol√≠gonos que delimitam uma √°rea, como bairros, munic√≠pios, estados ou pa√≠ses
    
    
- **Passo 2:** Construir uma **matriz de defasagens espaciais (ùêñ)**
    - O operador de **defasagem** tem a capacidade de criar novas vari√°veis a partir de outras j√° existentes
    - √â rara a utiliza√ß√£o de uma **matriz ùêñ** em sua forma bin√°ria, sugerindo a ado√ß√£o de algum um processo de **padroniza√ß√£o**, sendo os mais comuns a **padroniza√ß√£o da matriz ùêñ em linha, a dupla padroniza√ß√£o e a estabiliza√ß√£o da vari√¢ncia**


- **Passo 3::** Calcular as **autocorrela√ß√µes globais e locais**
    - Estabelecidas as **vizinhan√ßas**, e as suas respectivas **matrizes W** de **defasagem espacial**, podemos nos direcionar a verificar se os dados observados se distribuem de **forma aleat√≥ria** ou se formam um **padr√£o espacial**, ou seja, se existe a **autocorrela√ß√£o espacial** envolvendo o fen√¥meno estudado
    - A **autocorrela√ß√£o espacial** pode ser entendida como a medida da **correla√ß√£o** existente entre os valores de uma √∫nica vari√°vel de interesse de forma geogr√°fica
    - As **m√©tricas de autocorrela√ß√£o espacial global** direcionam-se a mensurar o grau de rela√ß√£o espacial de um fen√¥meno em rela√ß√£o a todos os valores observados na base de dados
    - As **m√©tricas de autocorrela√ß√£o espacial local**, medem as autocorrela√ß√µes das observa√ß√µes, uma a uma, em rela√ß√£o a sua vizinhan√ßa estabelecida pela matriz de defasagem espacial W
    
        **Autocorrela√ß√£o Global ‚Äì Diagrama da Estat√≠stica I de Moran:**
         - O **Diagrama de Moran** √© um m√©todo de visualiza√ß√£o das **autocorrela√ß√µes globais**, a t√©cnica visual consiste em um gr√°fico bidimensional de dispers√£o com quatro quadrantes, a saber: 
            - **Similaridade:** Alto-Alto (AA), Alto-Baixo (AB), **Dissimilaridade:** Baixo-Baixo (BB) e Baixo-Alto (BA)
    
      **Autocorrela√ß√£o Local ‚Äì a Estat√≠stica Moran Local:**
         - O :**√çndice de Moran Local (Ii):** identifica a rela√ß√£o existente entre um determinado pol√≠gono e a sua vizinhan√ßa, a partir de uma :**dist√¢ncia predefinida:**, por interm√©dio da :**covari√¢ncia:** existente entre eles, permitindo o exame da homogeneidade/diversidade dos dados
         - Os :**indicadores locais:** produzem um valor espec√≠fico para cada √°rea, permitindo, desse modo, a identifica√ß√£o de :**agrupamentos:**.



- **Passo 4:** Estima√ß√£o de modelos.

**Coluna 'DATAOCORRENCIA' precisa ser convertida para Datetime**

```python
#Converter a coluna DATAOCORRENCIA em Datetime

# Fun√ß√£o para extrair somente os primeiros numeros da vari√°vel de data
def extrai_data(x):
    return x[0:2]
# Extrai m√™s, semana e dia da coluna de data
dados["Dia"] = dados["DATAOCORRENCIA"].apply(extrai_data)

#Substituir a string '/' por '-' para transformar a vari√°vel em Datetime
dados['DATAOCORRENCIA'] = dados.DATAOCORRENCIA.replace('/','-')

# Transforamar a coluna de data Str em Datetime
dados['DATAOCORRENCIA']= pd.to_datetime(dados['DATAOCORRENCIA'], errors = 'coerce')
dados.info()
```

### Transformar as vari√°veis latitude e longitude em Geometry

```python
from shapely.geometry import Point

#Substituir ',' por '.' para transformar as vari√°veis em float
dados['LATITUDE'] = dados.LATITUDE.str.replace(',', '.')
dados['LONGITUDE'] = dados.LONGITUDE.str.replace(',','.')


#Transformar as vari√°veis Lat e Long em Int
dados['LATITUDE'] = pd.to_numeric(dados['LATITUDE'], errors='coerce')
dados['LONGITUDE'] = pd.to_numeric(dados['LONGITUDE'], errors='coerce')

# Criar a coluna Geometry 
geometry = [Point(xy) for xy in zip(dados['LONGITUDE'], dados['LATITUDE'])]


#Criar Geodataframe 
dados = gpd.GeoDataFrame(dados, geometry=geometry)
dados .head(5)
```
